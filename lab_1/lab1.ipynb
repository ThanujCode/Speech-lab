# A1. Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal

```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
```

```python
# printing the wave using librosa
y, sr = librosa.load('/Users/thanujbusa/Desktop/audio_sample.m4a')
librosa.display.waveshow(y)
```

# A2. Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to the spoken words and silence between the words.

```python
print("Length of the signal:",len(y),"samples")
print("Sampling rate:",sr, "Hz")
print("Magnitude range:",np.min(y),"to", np.max(y))
```

# A3. Take a small segment of the signal and play it.

```python
#dividing the signal into a segment
start = int(0.5 * sr)
end = int(2.5 * sr)
seg_wave = y[start:end]
librosa.display.waveshow(seg_wave)
ipd.Audio(seg_wave, rate=sr)
```

# A4. Play around with your recorded speech signal for various segments. Understand the nature of the signal. Also observe with abruptly segmented speech, how the perception of that speech is affected.

```python
#dividing the signal into 2 segments
seg1_start = int(1 * sr)
seg1_end = int(2.5 * sr)
seg2_start = int(2.8 * sr)
seg2_end = int(3.5 * sr)
seg1_wave = y[seg1_start:seg1_end]
seg2_wave = y[seg2_start:seg2_end]
```

```python
print("Playing Segment 1:")
librosa.display.waveshow(seg1_wave)
ipd.Audio(seg1_wave, rate=sr)
```

```python
print("Playing Segment 2:")
librosa.display.waveshow(seg2_wave)
ipd.Audio(seg2_wave, rate=sr)
```
